## creating bins and groups





When converting a continuous column to a categorical column, we can use three different pandas functionalities

1. `pd.cut()` - only applicable to convert continuous columns to categorical
2. `pd.qcut()` - only applicable for continuous columns
3. `np.where()` - only for both continuous and categorical columns

```python
stores = pd.read_csv('d:/data/stores.csv')
```

```python
stores
```

```python
## create groups of sales
## < 100 - low sales
## >= 100 and <= 200 - medium sales
## > 200 - high sales

stores['sales_groups'] = pd.cut(stores.TotalSales, bins=[0, 100, 200, 500])
```

```python
stores[['sales_groups', 'TotalSales']]
```

```python
## groups - unequal intervals
## bins - equal intervals
```

```python
stores['sales_bins'] = pd.cut(stores.TotalSales, bins=3, labels=['Low_sales', 'Medium_sales', 'High_sales' ])
```

```python
stores[['TotalSales', 'sales_bins']]
```

```python
## 0 - 100
## 100 - 300
## 300-500


pd.cut(stores.TotalSales, bins=[0, 100, 300, 500], labels=['G1', 'G2', 'G3'])
```

## `pd.qcut()`

```python
pd.qcut(stores.TotalSales, q=4, labels=['q1', 'q2', 'q3', 'q4'])
```

```python
pd.cut(stores.TotalSales, bins=4)
```

## Outlier treatment

What are outliers?

they are data points which are abnormally different from the rest of my data.


Why do we care?

Because they distort analysis.
Calculation of mean?
std?

What to do with them?

Clip them.

`Series.clip()`


```python
stores.describe(percentiles=[0.85, 0.9, 0.95, 0.99]).T
```

```python
## 1. Percentile

## 95%ile ,   5%ile
```

```python
lower_cutoff = stores.Total_Customers.quantile(0.05)
```

```python
high_cutoff = stores.Total_Customers.quantile(0.95)
```

```python
print(lower_cutoff)
print(high_cutoff)
```

```python
stores['TotalCustomersClipped'] = stores.Total_Customers.clip(upper=high_cutoff, lower=lower_cutoff)
```

```python
stores[['TotalCustomersClipped', 'TotalSales', 'Location', 'StoreType', 'Total_Customers']]
```

```python
high_cutoff
```

```python
lower_cutoff
```

```python
## p1p99
```

```python
## 2nd method

## Normal distribution

## lower_cutoff = mean - 3*std
## upper_cutoff = mean + 3*std



mean1 = stores.Total_Customers.mean()
std1 = stores.Total_Customers.std()

lower = mean1 - 3 * std1
upper = mean1 + 3 * std1

print(lower)
print(upper)
```

```python
stores.Total_Customers.clip(upper=upper, lower=lower)
```

```python
## IQR method

## q3, q1

q1 = stores.Total_Customers.quantile(0.25)
q3 = stores.Total_Customers.quantile(0.75)

iqr = q3 - q1


lower2 = q1 - 1.5*iqr
upper2 = q3 + 1.5*iqr

print(lower2)
print(upper2)
```

```python
stores.Total_Customers.clip(upper=upper2, lower=lower2)
```

```python
def treat_outliers(x):
    
    q1 = x.quantile(0.25)
    q3 = x.quantile(0.75)

    iqr = q3 - q1


    lower2 = q1 - 1.5*iqr
    upper2 = q3 + 1.5*iqr

    return x.clip(lower=lower2, upper=upper2)
```

## applymap and apply in pandas


`.apply(fn_name)`

this fn_name again can be anything:

1. user defined function
2. system defined function
3. lambda function

for DataFrame - this will take one Series at a time and will pass it to apply()
for Series - this will take one value at a time from the Series and will pass it to apply()


`.applymap()`

only for DataFrame - this will work for each cell of the DataFrame


alternate to running a for but much more efficient.

```python
stores.Location.apply(lambda x: x.upper())
```

```python
def upper_case(x):
    return x.upper()
```

```python
stores.Location.apply(upper_case)
```

```python
## treating outliers
```

```python
## one question

## Q: create a function that takes in a series and calculates the aggregates of it
        ## .min(), .max(), .mean()
    
## Q: apply this function on the entire continuous vars dataframe
```

```python
def calculate_aggregates(series):
    

    aggregates = {
        'Minimum': series.min(),
        'Maximum': series.max(),
        'Mean': series.mean(),
        'Median': series.median(),
        'Sum': series.sum(),
        'Count': series.count(),
        'Standard Deviation': series.std(),
    }
    return pd.Series(aggregates)
```

```python
stores.select_dtypes('number').drop(columns='Dummy').apply(calculate_aggregates).T
```
