## Accessing columns or subsetting

There are 4 ways of going about it.

1. `[]`

2. `.loc[]`

3. `.iloc[]`

    can't be used to create new columns

4. `df.column_name`
    limitations:
        will only be used to extract one column at a time
        
        also column names have to follow INC rules to be able to use this syntax
        
        also can't be used to create new columns

## for boolean indexing, use

`.loc[]` - works for both rows and columns

`[]` - work on only the rows

can't use .iloc for boolean indexing

```python
## Q: Get all the records where OPerating Cost is less 30 and greater than 20

(data1.OperatingCost > 20) & (data1.OperatingCost < 30)
```

```python
data1.loc[(data1.OperatingCost > 20) & (data1.OperatingCost < 30)].head(2)
```

```python
data1.iloc[data1.Location == 'Delhi']
```

## `.isin()`       `.between()`

```python
## ## Q: Get all the records where OPerating Cost is less 30 and greater than 20


data1.OperatingCost.between(20, 30).head(2)
```

```python
## Q: get the stores from Delhi and Mumbai

data1.Location.isin(['Delhi', 'Mumbai'])
```
