```python
# null values will have to be dealt
# Interest rate, Employment length is of object dtype
    ## to be able to change these data types, we will need to clean % 
    ## cleaning for Employement length
    
## create a new loan_length integer
## retain the orginal loan lenght in category

## renaming


```

```python
data2['Loan.Length'].unique()
```

```python
def wrangle_data(df):
    
    ## renaming columns
    
    df.columns = [column.replace('.', '') for column in df.columns]
    
    
    ## Interest rate cleaning
    
    df['InterestRate'] = df.InterestRate.str.replace('%', '')
    
    ## Loan Length
    
    df['LoanLength'] = df.LoanLength.str.replace(' months', '')
    
    ## Employment lenght
    
    ## Debt To Income Rate
    
    df['DebtToIncomeRatio'] = df.DebtToIncomeRatio.str.replace('%', '')
    
    df['EmploymentLength'] = ( df.EmploymentLength.str.replace('<', '')
                                    .str.replace('years', '').str.replace('+', '')
                             .str.replace('year', ''))
    
    return df
```

```python
clean_data2 = wrangle_data(data2)
```
