```python
## take a random sample of 20 observations from this population
pop[np.random.randint(0,10000, 20)]
```

```python
## Plot the sampling distribution of sample means take out of the above
## population for different sample sizes (n)


## initial sample size of 1
n = 1

## subplots object with 6 rows and 5 columns of subplots
fig, ax = plt.subplots(6, 5, figsize=(20, 8))

## loop over each subplot
for r in range(6):  
    for c in range(5):

        ## for each plot define an empty list that we will contain all the sample means
        sample_means = []
        
        ## take 1000 samples out of the population for the given sample size, n
        for i in range(1000):
            sample_mean = pop[np.random.randint(0, 10000, n)].mean()
            sample_means.append(sample_mean)

        sns.histplot(sample_means, ax=ax[r, c])
        
        n += 1
```
