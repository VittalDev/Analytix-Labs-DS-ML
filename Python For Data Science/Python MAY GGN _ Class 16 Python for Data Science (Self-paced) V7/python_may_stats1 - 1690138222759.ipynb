### Normal Distribution

It is a specific type of probability distribution that follows a bell-shaped curve. It is the most important distribution
that we learn in statistics as normal distributions can be found all around in nature. Many kinds of data follows normal
distribution including stock price returns, heights of people, etc. Why normal distributions are found all around in nature also has an explanation that will be talked later when we talk about the Central Limit Theorem.




#### Properties of Normal Distribution

1. Bell-shaped curve
2. It is always symmetric around the centre
3. Mean = Median = Mode
4. It is defined completely only by the mean of the data and the standard deviation of the data`
